**Q1. Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS)**

- **Artificial Intelligence (AI):**
Broad field focused on building systems that can mimic human intelligence (e.g., reasoning, problem-solving, decision-making). Example: Chatbots, self-driving cars.

- **Machine Learning (ML):**
Subset of AI that enables systems to learn patterns from data and improve without being explicitly programmed. Example: Spam email detection.

- **Deep Learning (DL):**
Subset of ML using artificial neural networks with multiple layers (deep networks). Excellent for handling unstructured data like images, text, and speech. Example: Face recognition in photos.

- **Data Science (DS):**
Interdisciplinary field combining statistics, ML, and domain knowledge to extract insights from data. It covers data collection, cleaning, visualization, and predictive modeling. Example: Customer churn analysis in telecom.

**Q2. What are the types of machine learning? Describe each with one real-world example.**

There are five type of machine learning:

**1. Supervised Learning:** The algorithm is trained on labeled data to develop and optimize machine learning models. This model needs to map the input to the output. This learning is mostly used for image recognition, fraud detection, and risk assessments.

- **Real-world Example:**
Credit card fraud detection. A model can be trained on historical transaction data that is labeled as either "fraudulent" or "legitimate," allowing it to identify new transactions that are likely to be fraudulent.

**2. Unsupervised Learning:** This model uses unlabelled data to train machines. There is no output variable. Instead, the variable learns from the data, finds patterns and features in the data, and returns the output.

- **Real-world Example:**
Customer segmentation. A retail company can use unsupervised learning to analyze customer data and automatically group customers into different segments based on their purchasing behaviors or demographics, without having pre-defined customer groups.

**3. Semi-supervised Learning:** It is a combination of supervised and unsupervised learning techniques. The dataset has both labeled and unlabelled data. An example of semi-supervised learning is Generative Adversarial Networks (GANs), a deep learning technique that trains two neural networks and generates unlabeled data.

- **Real-world Example:**
Medical image analysis. A small set of medically labeled images can help a model learn to classify a much larger collection of unlabeled medical images, improving diagnostic accuracy.

**4. Reinforcement Learning:** Reinforcement learning trains a machine to take suitable actions and maximize reward in a particular situation. It uses an agent and an environment to produce actions and rewards.

- **Real-world Example:**
Robotic navigation. Robots can be trained using reinforcement learning to navigate complex environments, learn to grasp objects, or move efficiently by receiving positive reinforcement for successful movements and negative feedback for failures

**5. Self-supervised Learning:** Unlike other models that require massive labeled datasets, this model is trained on unlabelled data. It transforms unstructured datasets into structured ones.

- **Real-world Example:**
Natural Language Processing (NLP). Large language models can use self-supervised learning by predicting the next word in a sentence, which helps them understand grammar, context, and meaning without needing explicit human labeling for every sentence.

**Q3. Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.**

**1. Overfitting**

**Definition:**
Overfitting happens when a machine learning model learns the training data too well, including noise and random fluctuations, instead of just the underlying pattern.

**Characteristics:**

- Very high accuracy on training data.

- Poor performance on test or unseen data (low generalization).

- Model is too complex (too many parameters, deep decision trees, etc.).

**Example:**
A decision tree that grows very deep and creates a separate rule for almost every training example. While it predicts training data perfectly, it fails to predict new data accurately.


**2. Underfitting**

**Definition:**
Underfitting happens when a model is too simple and fails to capture the important patterns in the data.

**Characteristics:**

- Low accuracy on both training and test data.

- Model has high bias (too many simplifying assumptions).

**Example:**
Using a straight line (linear regression) to predict data that follows a curve. The line is too simple to capture the real relationship.



**3. Bias-Variance Tradeoff**

**Bias:** Error due to overly simple assumptions in the model. High bias leads to underfitting.

**Variance:** Error due to modelâ€™s sensitivity to small fluctuations in the training data. High variance leads to overfitting.

**Tradeoff:**

- Increasing model complexity decreases bias but increases variance.

- Decreasing model complexity decreases variance but increases bias.

- The goal is to find the optimal balance where both bias and variance are reasonably low, achieving good performance on both training and test data.

**Q4. What are outliers in a dataset, and list three common techniques for handling them.**


Outliers are data points that are significantly different from other data points in a dataset. They can occur due to measurement errors, data entry errors, or genuine but unusual observations. Outliers can skew statistical analyses and negatively impact the performance of machine learning models.


**Techniques for Handling Outliers:**



**1.Deletion (Trimming):** Removing outliers from the dataset. This is suitable if the outliers are due to errors or do not represent the true population. However, it can lead to loss of valuable data if the outliers are genuine.
- **Example:** If you have a dataset of human heights and one entry is 500 cm, it's likely a data entry error and can be removed.

**2.Transformation:** Applying mathematical transformations to the data to reduce the impact of outliers. Common transformations include logarithmic, square root, or Box-Cox transformations.
- **Example:** Using a logarithmic transformation on skewed income data can compress the range of large values, making outliers less influential.

**3.Winsorizing (Capping):** Limiting extreme values in the dataset to a certain percentile (e.g., replacing all values above the 95th percentile with the value at the 95th percentile).

- **Example:** In a dataset of test scores, you might cap all scores above 98% at 98% to reduce the influence of extremely high scores.

**Q.5 Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.**


**Process of handling missing values:**

**1. Identification:** The first step is to identify missing values in the dataset. This can be done by inspecting the data, using methods like `.isnull().sum()` in pandas to count missing values per column, or visualizing missing data patterns.

**2.Understanding the Cause:** It's important to try and understand why the data is missing. Is it random, due to a systematic error, or because of a specific reason? This understanding can influence the chosen handling technique.

**3.Choosing a Technique:** Based on the extent of missing data, the type of data (numerical or categorical), and the potential impact on analysis, a suitable technique for handling missing values is chosen.

**4.Implementation:** Apply the chosen technique to the dataset.



**Techniques:**

**Deletion:**
- **Row Deletion:** Removing entire rows with missing values. This is suitable if only a small percentage of rows have missing data.
- **Column Deletion:** Removing entire columns if a large percentage of values are missing in that column.
- **Imputation:** Replacing missing values with estimated values.


**Imputation Techniques:**

*   **For Numerical Data:**
    *   **Mean/Median Imputation:** Replacing missing values with the mean or median of the column. The median is often preferred for skewed data as it is less affected by outliers.

    *   **Example:** If the 'Age' column has missing values, you could replace them with the median age of all individuals in the dataset.

*   **For Categorical Data:**
    *   **Mode Imputation:** Replacing missing values with the mode (most frequent category) of the column.

    *   **Example:** If the 'City' column has missing values, you could replace them with the most frequently occurring city in the dataset.

```python
# Answer-8

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Generate 1000 samples from a normal distribution
data = np.random.randn(1000)

# 2. Introduce 50 random missing values
num_missing = 50
missing_indices = np.random.choice(1000, num_missing, replace=False)
data[missing_indices] = np.nan

# 3. Create a copy of the original data for plotting
data_before = data.copy()

# Convert the NumPy array to a pandas DataFrame for easier handling of missing values
df = pd.DataFrame(data, columns=['Value'])

# 4. Fill missing values with the column mean
# Calculate the mean of the non-missing values
mean_value = df['Value'].mean()

# Fill the missing values with the calculated mean
df['Value'].fillna(mean_value, inplace=True)

# 5. Plot a histogram before and after imputation
plt.figure(figsize=(12, 6))

# Histogram before imputation
plt.subplot(1, 2, 1)
plt.hist(data_before[~np.isnan(data_before)], bins=30, alpha=0.7, color='blue')
plt.title("Before Imputation (Original Data)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Histogram after imputation
plt.subplot(1, 2, 2)
plt.hist(df['Value'], bins=30, alpha=0.7, color='orange')
plt.title("After Imputation (Mean Imputation)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

print("\nNumber of missing values before imputation:", np.isnan(data_before).sum())
print("Number of missing values after imputation:", df['Value'].isnull().sum())
print("Calculated mean for imputation:", mean_value)
```

**1. Assess missingness:**

Determine if the data is "missing completely at random," "missing at random," or "missing not at random" by examining the relationship between missingness in age and other variables.

**2. Choose an imputation strategy:**
- Regression imputation: Build a machine learning model (e.g., linear regression or a tree-based model) to predict missing age values based on other features like transaction_amount or payment_method. This preserves relationships between variables and is a robust choice.

**Step 3: Address Outliers**

Outliers in transaction_amount can skew your model, especially linear ones. Since extreme transaction amounts can be legitimate (e.g., high-value purchases) or fraudulent, simply deleting them is not always the best approach.

**1. Detect outliers:**

Use statistical methods like the Interquartile Range (IQR) or z-scores to identify data points that fall outside the normal range.

**2. Choose an outlier treatment:**
- **Winsorization (Capping):** Cap the outliers by replacing them with a less extreme value, such as the value at the 95th or 99th percentile. This is a good option if you believe the outliers are valid but too influential.
- **Log transformation:** For a right-skewed transaction_amount variable, applying a log transformation can normalize the distribution, which reduces the effect of outliers.
- **Create a new feature:** In a fraud detection context, the transaction amount itself might be highly informative. Consider creating a binary feature that flags whether a transaction is an outlier, allowing the model to learn from this distinction


**Step 4: Handle Class Imbalance**

A highly imbalanced dataset can cause a model to become biased towards the majority class (non-fraud), resulting in poor performance on the minority class (fraud). This means high accuracy but low recall for fraud cases.

**1.Re-sampling techniques:**
- **Oversampling:** Increase the number of instances in the minority class. The most popular method is SMOTE (Synthetic Minority Over-sampling Technique), which generates synthetic data points for the minority class rather than simply duplicating existing ones.
- **Undersampling:** Decrease the number of instances in the majority class to match the minority class. A disadvantage is the potential loss of valuable information from the majority class.
- **Hybrid methods:** Use a combination of oversampling and undersampling, such as SMOTE with Edited Nearest Neighbors (ENN).

**2.Algorithm-level approaches:**

- **Class weights:** Many machine learning algorithms (like tree-based methods and logistic regression) allow you to specify class weights. You can set a higher weight for the minority class to penalize the model more for misclassifying fraud.
- **Change evaluation metrics:** Do not rely on accuracy for imbalanced datasets. Instead, focus on metrics like precision, recall, F1-score, and the Area Under the Receiver Operating Characteristic (ROC-AUC) curve, which provide a more accurate picture of performance.

**Step 5: Encode Categorical Variables**

Machine learning models cannot process text directly, so categorical variables like payment_method must be converted into numerical format.

**1.Choose an encoding method:**

- **One-Hot Encoding:** Best for nominal (unordered) categorical variables. This method creates a new binary column for each category. For example, payment_method with categories "credit card," "debit card," and "bank transfer" would become three new binary columns.
- **Target Encoding:** Replaces each category with a blend of the mean target value for that category. This is particularly useful for high-cardinality categorical features and may be effective for fraud detection. You must use cross-validation to prevent overfitting
